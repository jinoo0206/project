```python
import cv2
import numpy as np
import sys
```

```python
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./haarcascade_eye.xml')
```

```python
img = cv2.imread('./fig/face_2.png')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(img_gray, 1.3, 3)

for (x,y,w,h) in faces:
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (225,0,0),2)
    roi_g = img_gray[y:y+h, x:x+h]
    roi_c = img[y:y+h, x:x+h]
    
    eyes = eye_cascade.detectMultiScale(roi_g)
    
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_c, (ex,ey), (ex+ew, ey+eh), (0,225,0),2)
        
cv2.imshow('img',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)


tony = cv2.imread('./fig/mask.JPG')

tony_gray = cv2.cvtColor(tony, cv2.COLOR_BGR2GRAY)
retval, tony_mask = cv2.threshold(tony_gray, 240,255,cv2.THRESH_BINARY_INV)

if not cap.isOpened:
    print('video read failed')
    sys.exit()
    
while True:
    
    ret, img = cap.read()
    
    if ret is None:
        print('frame read failed')
        break
        
    original_tony_h, original_tony_w, tony_channals = tony.shape
    img_h , img_w, img_channels = img.shape
        
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(img_gray, 1.3, 3)    
    
    for x,y,w,h in faces:
    
        tony_w = int(w)
        tony_x = int(x)
        tony_h = int(h)
        tony_y = int(y)


        tony_x1 =  int(x )
        tony_y1 =  int(y )
        tony_x2 = tony_x1 + tony_w
        tony_y2 = tony_y1 + tony_h

        if tony_x1 < 0:
            tony_x1 = 0
        if tony_y1 < 0:
            tony_y1 = 0
        if tony_x2 > img_w:
            tony_x2 = img_w
        if tony_y2 > img_h:
            tony_y2 = img_h

        tony = cv2.resize(tony, (tony_w, tony_h), interpolation = cv2.INTER_AREA)
        mask = cv2.resize(tony_mask, (tony_w, tony_h), interpolation = cv2.INTER_AREA)
        mask_inv = cv2.resize(original_mask_inv, (tony_w, tony_h), interpolation = cv2.INTER_AREA)

        roi = img[tony_y : tony_y + tony_h, tony_x : tony_x + tony_w]

        roi_bg = cv2.bitwise_and(roi,roi,mask = mask_inv)
        roi_fg = cv2.bitwise_and(tony,tony,mask=mask)
        dst = cv2.add(roi_bg,roi_fg)

        img[tony_y1 : tony_y2, tony_x1 : tony_x2] = dst
    
    
    
    
    cv2.imshow('frame', img)
    
    key = cv2.waitKey(20)
    
    if key == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
